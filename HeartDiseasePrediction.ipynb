```python
#As svc is not so efficient,  we will use grid search cv to improve efficiency.
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.00001],'kernel': ['rbf']}
```

```python
grid = GridSearchCV(SVC(),param,verbose=3)
```

```python
x = df.drop('Heart Disease',axis=1)
y = df['Heart Disease']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
grid_pred = grid.predict(x_test)
```

```python
print(confusion_matrix(grid_pred,y_test))
```

```python
print(classification_report(grid_pred,y_test))
```
