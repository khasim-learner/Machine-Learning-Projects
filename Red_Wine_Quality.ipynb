Data Preprocessing

```python
x = df.drop(columns='quality',axis=1)
```

```python
x
```

Label Binarisation - Means there are 6 values for quality, but i gonna change into 2 values. Quality of 7 and 8 are good. Quality below 7 are bad.

```python
y = df['quality'].apply(lambda y_val:1 if y_val>=7 else 0)
```

```python
y
```

Train and Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

Model Training :
Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

Model Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
#Accuracy on Test data

x_test_pred = model.predict(x_test)
test_data_accur = accuracy_score(x_test_pred,y_test)
```

```python
test_data_accur
```

Building a Predicitve System

```python
def make_prediction(input_data, model):
    # Convert input data to numpy array
    inp_numpy = np.asarray(input_data)

    # Reshape the data as we only want result for one instance
    inp_np_shape = inp_numpy.reshape(1, -1)

    # Make prediction
    prediction = model.predict(inp_np_shape)

    if prediction == 1:
        return "Good quality wine"
    else:
        return "Bad quality wine"
```

```python
input_data = (7.3, 0.65, 0.0, 1.2, 0.065, 15.0, 21.0, 0.9946, 3.39, 0.47, 10.0)


prediction = make_prediction(input_data, model)
print(prediction)
```

```python
input_data = (7.8,0.645,0.0,2.0,0.08199999999999999,8.0,16.0,0.9964,3.38,0.59,9.8)


prediction = make_prediction(input_data, model)
print(prediction)
```
