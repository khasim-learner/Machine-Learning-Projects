Handling outliers using Inter quartile range

```python
import numpy as np

def detect_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers in {col}: {outliers.shape[0]} rows")

for col in ['temperature_celsius', 'humidity', 'pressure_mb', 'wind_kph', 'precip_mm']:
    detect_outliers(col)
```

Convert Categorical Variables to Numerical

```python
# Define function to group weather conditions
def categorize_weather(condition):
    condition = condition.lower()  # Convert to lowercase for consistency

    if "sunny" in condition or "clear" in condition:
        return "Clear"
    elif "cloudy" in condition or "mist" in condition or "fog" in condition:
        return "Cloudy"
    elif "rain" in condition or "drizzle" in condition:
        return "Rain"
    elif "snow" in condition or "blizzard" in condition:
        return "Snow"
    elif "thunder" in condition:
        return "Thunderstorm"
    elif "sleet" in condition or "freezing" in condition:
        return "Sleet"
    else:
        return "Other"  # Unclassified cases

# Apply the function
df['condition_category'] = df['condition_text'].apply(categorize_weather)

```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['condition_category'] = le.fit_transform(df['condition_category'])
```

Advanced EDA
Anomaly detection

```python
from scipy.stats import zscore

# Select relevant numerical columns
cols = ['temperature_celsius', 'precip_mm', 'wind_kph', 'air_quality_PM2.5']
df_selected = df[cols]

# Compute Z-scores
z_scores = df_selected.apply(zscore)

# Define threshold (|Z| > 3 is an outlier)
threshold = 3
outliers = (z_scores.abs() > threshold)

# Count outliers per column
print(outliers.sum())

# Filter outlier rows
df_outliers = df[outliers.any(axis=1)]
df_outliers.head()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_selected)
plt.xticks(rotation=45)
plt.title("Boxplot of Weather Features (Outlier Detection)")
plt.show()
```

ForeCasting with Multiple Models

```python
import pandas as pd
df2 = pd.read_csv('/content/GlobalWeatherRepository.csv')
```

```python
df2['last_updated'] = pd.to_datetime(df2['last_updated'])
df2.set_index('last_updated', inplace=True)

# Select target variable (e.g., temperature)
df_forecast = df2[['temperature_celsius']]
```

```python
from sklearn.model_selection import train_test_split

train_size = int(len(df_forecast) * 0.8)
train, test = df_forecast[:train_size], df_forecast[train_size:]
```

1️⃣ ARIMA Model (AutoRegressive Integrated Moving Average)

```python
from statsmodels.tsa.arima.model import ARIMA

model_arima = ARIMA(train, order=(5,1,0))  # (p,d,q) parameters
arima_fit = model_arima.fit()
forecast_arima = arima_fit.forecast(steps=len(test))

# Plot ARIMA Forecast
plt.figure(figsize=(10, 5))
plt.plot(train, label="Train")
plt.plot(test, label="Actual")
plt.plot(test.index, forecast_arima, label="ARIMA Forecast", linestyle="dashed")
plt.legend()
plt.title("ARIMA Temperature Forecast")
plt.show()
```

2.  LSTM (Deep Learning Approach)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

# Prepare Data for LSTM
train_values = np.array(train).reshape(-1, 1)
test_values = np.array(test).reshape(-1, 1)

# Define Model
model_lstm = Sequential([
    LSTM(50, activation='relu', return_sequences=True, input_shape=(train_values.shape[1], 1)),
    LSTM(50, activation='relu'),
    Dense(1)
])
model_lstm.compile(optimizer='adam', loss='mse')

# Train Model
model_lstm.fit(train_values, train_values, epochs=20, verbose=1)

# Forecast
forecast_lstm = model_lstm.predict(test_values)
plt.plot(test.index, forecast_lstm, label="LSTM Forecast", linestyle="dashed")
plt.legend()
plt.title("LSTM Temperature Forecast")
plt.show()
```

 Create an Ensemble Model for Better Accuracy

```python
import numpy as np
import matplotlib.pyplot as plt


# Generate forecasts using ARIMA & LSTM models
forecast_arima = arima_fit.forecast(steps=len(test))
forecast_lstm = model_lstm.predict(test_values)

# Ensure both forecasts have the same shape
forecast_lstm = forecast_lstm.flatten()

# ---------------- Ensemble Forecast ----------------
# Weighted Average of ARIMA & LSTM
ensemble_forecast = (0.5 * forecast_arima) + (0.5 * forecast_lstm)  # Adjust weights if needed

# ---------------- Plot the Results ----------------
plt.figure(figsize=(10, 5))
plt.plot(test.index, test, label="Actual")
plt.plot(test.index, forecast_arima, label="ARIMA Forecast", linestyle="dashed")
plt.plot(test.index, forecast_lstm, label="LSTM Forecast", linestyle="dashed")
plt.plot(test.index, ensemble_forecast, label="Ensemble Forecast", linestyle="dashed", color="black")

plt.legend()
plt.title("Ensemble Model (ARIMA + LSTM) Temperature Forecast")
plt.show()
```
